### Simple Recomendation

```python
vote_data=meta[['original_title','vote_average','vote_count','image']]
vote_data['vote_average'].mean()
```

#### Create new Score using Weight Rating

```python
def weight_rating(row):
    v=row['vote_count'] # v is the number of votes for the movie
    m=1500 # minimal vote count yang ingin di tampilkan di chat 'm is the minimum votes required to be listed in the chart
    r=row['vote_average'] #R is the average rating of the movie
    c=5.6 # mean vote_average C is the mean vote across the whole report
    wr=((v/(v+m)*r)+(m/(v+m)*c))
    return wr
vote_data['score'] = vote_data.apply(weight_rating,axis=1)
vote_data.head(10)
```

```python
def Show_Movie_by_vote_and_rating(X):
    Byscore=vote_data.sort_values(by='score',ascending=False).head(X)[['original_title','score','image']]
    Byvote=vote_data.sort_values(by='vote_count',ascending=False).head(X)[['original_title','vote_count','image']]
    return [Byscore,Byvote]
print(Show_Movie_by_vote_and_rating(10)[0])
print(Show_Movie_by_vote_and_rating(10)[1])
```

### Content Based

#### Genre Based

```python
from ast import literal_eval
content_based_genres=meta[meta['vote_count']>250][['original_title','genres','id','image']].drop_duplicates()
content_based_genres.reset_index(drop=True,inplace=True)
def newgenres(x):
    genres=[]
    x=literal_eval(x)
    for item in x:
        genres.append(item['name'])
    return ' '.join(genres)
content_based_genres['newgenres']=content_based_genres['genres'].apply(newgenres)
content_based_genres.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
cv_genres_result = cv.fit_transform(content_based_genres['newgenres'])
cv_genre_df=pd.DataFrame(cv_genres_result.todense(),columns=cv.get_feature_names(),index=content_based_genres['original_title'])
cv_genre_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
con_sim_movie= cosine_similarity(cv_genres_result)
index_to_search=content_based_genres[content_based_genres['original_title']=='Jumanji'].index[0]
pd.Series(con_sim_movie[index_to_search],index=content_based_genres['original_title']).sort_values(ascending=False).head(10)
```

```python
def get_recomendation_based_genres(title):
    index_to_search=content_based_genres[content_based_genres['original_title']==title].index[0]
    series_similiar=pd.Series(con_sim_movie[index_to_search])
    index_similiar = series_similiar.sort_values(ascending=False).head(10).index
    return content_based_genres.loc[index_similiar]
get_recomendation_based_genres('Kick-Ass')
```

#### Cast Based

```python
content_based_genres
```

```python
content_based_genres_cast_crew=meta[meta['vote_count']>250][['original_title','genres','id','image']]
credits=pd.read_csv('the-movies-dataset/credits.csv')
content_based_genres['id']=pd.to_numeric(content_based_genres['id'])
cb_genres_cast_crew=content_based_genres.merge(credits,on=['id'])
cb_genres_cast_crew=cb_genres_cast_crew.drop_duplicates('id')
cb_genres_cast_crew.reset_index(drop=True,inplace=True)
```

```python
def new_cast(x):
    casts = []
    x = literal_eval(x)
    for item in x:
        name = item['name'].replace(' ','')
        casts.append(name)
    return ' '.join(casts)
cb_genres_cast_crew['new_crew'] = cb_genres_cast_crew['crew'].apply(new_cast)
cb_genres_cast_crew['new_cast'] = cb_genres_cast_crew['cast'].apply(new_cast)
cb_genres_cast_crew.head()
```

```python
cv_res_cast = CountVectorizer().fit_transform(cb_genres_cast_crew['new_cast'])
cos_sim_cast = cosine_similarity(cv_res_cast)
def get_recommendation_based_cast(title):
    index_to_search = cb_genres_cast_crew[cb_genres_cast_crew['original_title'] == title].index[0]
    series_similar = pd.Series(cos_sim_cast[index_to_search])
    index_similar = series_similar.sort_values(ascending=False).head(10).index
    return cb_genres_cast_crew.loc[index_similar][['original_title','new_cast']]
get_recommendation_based_cast('The Avengers')
```

#### Crew Based

```python
def new_directed(x):
    x = literal_eval(x)
    for item in x:
        if (item['job'] == 'Director') :
            return (item['name'].replace(' ',''))
cb_genres_cast_crew['new_crew'] = cb_genres_cast_crew['crew'].apply(new_directed)
cv_res_crew = CountVectorizer().fit_transform(cb_genres_cast_crew['new_crew'])
cos_sim_crew = cosine_similarity(cv_res_crew)
cb_genres_cast_crew.rename(columns = {'new_crew':'Director'},inplace=True)
```

```python
def get_recommendation_based_crew(title):
    index_to_search = cb_genres_cast_crew[cb_genres_cast_crew['original_title'] == title].index[0]
    series_similar = pd.Series(cos_sim_crew[index_to_search])
    index_similar = series_similar.sort_values(ascending=False).head(10).index
    return cb_genres_cast_crew.loc[index_similar][['original_title','new_cast','Director']]
get_recommendation_based_crew('Iron Man')
```

```python
user_movie_pivot=user_movie.pivot_table(index='original_title',columns='userId',values='rating')
user_movie_pivot.dropna(thresh=10,inplace=True)
user_movie_pivot.fillna(0,inplace=True)
cos_sim_ratings=cosine_similarity(user_movie_pivot)
df_sim_rating=pd.DataFrame(cos_sim_ratings,index=user_movie_pivot.index,columns=user_movie_pivot.index)
def get_recomended_based_rating(movie):
    list_Ser=[]
    for item in movie:
        list_Ser.append(df_sim_rating[item[0]]*(item[1]-2.5))
    return sum(list_Ser).sort_values(ascending=False).head(10)
```

```python
def get_recomendation_user_based(ratings):
    userpivot1=userpivot.copy()
    movie_index=np.array(ratings)[0:,0]
    movie_rating=np.array(ratings)[0:,1]
    userpivot1.loc[(userpivot1.index[-1]+1),movie_index] = movie_rating
    userpivot1.fillna(0,inplace=True)
    similiarity= cosine_similarity(userpivot1)
    user_id_sim=pd.DataFrame(similiarity,columns=userpivot1.index,
                             index=userpivot1.index).loc[userpivot1.index[-1]].sort_values(ascending=False)[1:11].index
    ser=[]
    for id in user_id_sim:
        movie_rating=userpivot1.loc[id]
        ser.append(movie_rating)
    return (sum(ser)/10).sort_values(ascending=False).drop(movie_index).head(10)
get_recomendation_user_based([['Armageddon',5],['Harry Potter and the Prisoner of Azkaban',4],['Hulk',2]])
```
