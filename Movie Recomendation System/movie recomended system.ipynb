## SIMPLE RECOMENDED

```python
vote_data=meta[['original_title','vote_average','vote_count']]
```

```python
vote_data['vote_average'].mean()
```

```python
def weight_rating(row):
    v=row['vote_count'] # v is the number of votes for the movie
    m=1500 # minimal vote count yang ingin di tampilkan di chat 'm is the minimum votes required to be listed in the chart
    r=row['vote_average'] #R is the average rating of the movie
    c=5.6 # mean vote_average C is the mean vote across the whole report
    wr=((v/(v+m)*r)+(m/(v+m)*c))
    return wr
```

```python
vote_data['score'] = vote_data.apply(weight_rating,axis=1)
vote_data
```

```python
vote_data.sort_values(by='score',ascending=False)
```

```python
vote_data.sort_values(by='vote_count',ascending=False)
```

```python
def newgenres(x):
    genres=[]
    x=literal_eval(x)
    for item in x:
        genres.append(item['name'])
    return ' '.join(genres)
```

```python
def new_cast(x):
    casts = []
    x = literal_eval(x)
    for item in x:
        name = item['name'].replace(' ','')
        casts.append(name)
    return ' '.join(casts)
```

```python
def get_recomended_based_rating(movie):
    list_Ser=[]
    for item in movie:
        list_Ser.append(df_sim_rating[item[0]]*(item[1]-2.5))
    return sum(list_Ser).sort_values(ascending=False).head(10)
```

```python
def get_recomendation_user_based(ratings):
    movie_index=np.array(ratings)[0:,0]
    movie_rating=np.array(ratings)[0:,1]
    userpivot.loc[(userpivot.index[-1]+1),movie_index] = movie_rating
    userpivot.fillna(0,inplace=True)
    similiarity= cosine_similarity(userpivot)
    user_id_sim=pd.DataFrame(similiarity,columns=userpivot.index,
                             index=userpivot.index).loc[userpivot.index[-1]].sort_values(ascending=False)[1:11].index
    ser=[]
    for id in user_id_sim:
        movie_rating=userpivot.loc[id]
        ser.append(movie_rating)
    return (sum(ser)/10).sort_values(ascending=False).drop(movie_index)
```
