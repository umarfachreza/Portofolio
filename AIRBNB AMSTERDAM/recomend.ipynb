```python
def removedollar(price):
    if ',' in str(price) :
        str_price = str(price).replace(",", "")
        return str_price.split('.')[0][1:]
    else :
        return str(price).split('.')[0][1:]
new_listing['price'] = pd.to_numeric(new_listing['price'].apply(removedollar))
```

```python
new_listing.dropna(inplace=True)
new_listing['listing_id']=new_listing['id']
new_listing.drop(['id'], axis=1,inplace=True)
new_listing.info()
```

```python
new_listing=new_listing.reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re
def clean_text(x):
    letter = re.sub("[^a-zA-Z1-9]","_",x)
    return ' '.join(letter.split()).lower()

def new_amenties(x):
    fasilitas = str(x)[1:-1].split(',')
    fas=[]
    for item in fasilitas:
        name = clean_text(item)
        if 'missing' not in name :
            fas.append(name)
    return ' '.join(fas)
new_listing['amenities']=new_listing['amenities'].apply(new_amenties)
```

```python
cv_res = CountVectorizer()
cv_res_hotel = cv_res.fit_transform(new_listing['amenities'])
cos_sim = cosine_similarity(cv_res_hotel)
```

```python
df_sim=pd.DataFrame(cos_sim,index=new_listing.index,columns=new_listing.listing_id)
```

```python
df_sim['loc'] = new_listing['neighbourhood_cleansed']
df_sim['loc']
```
